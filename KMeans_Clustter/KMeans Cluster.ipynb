```python
# kmeans algorithm : unsupervised learning(clustering)

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

# step : 1 load the data
plt.figure()
data = pd.read_csv("Book1.csv")
x = data.iloc[:,0]
y = data.iloc[:,1]
plt.scatter(x,y)

# step : 2 load the model
model=KMeans(init="k-means++",n_clusters=4)

# step : 3 train the model
model.fit(data)

# use the model to diaplay the clusters
result = model.predict(data)
print(result)

# find the centroids (center point)
centroids = model.cluster_centers_
print(centroids)
```

```python
# Load the CSV file
data = pd.read_csv("clustering.csv")
x = data["ApplicantIncome"]
y = data["LoanAmount"]
data = list(zip(x,y))

# Plot the dataset to identify the clusters
plt.figure()
plt.scatter(x, y, marker='o', facecolors='none', edgecolors='k', s=30)
plt.title('Input data')
plt.xticks(())
plt.yticks(())

# Load the Model
kmeans = KMeans(init='k-means++', n_clusters=3)

# Train the Model
kmeans.fit(data)

# Use the Model to predict values
results = kmeans.predict(data)
print(results)

# Find the centroids
centroids = kmeans.cluster_centers_
print(centroids)

# Plot the clusters along with centroids
plt.figure()
plt.scatter(x, y, marker='o', facecolors='none', edgecolors='k', s=30)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=200, color='r')
plt.title('Input data')
plt.xticks(())
plt.yticks(())
```

```python
import numpy as np

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

# Load the CSV file
data = pd.read_csv("Wholesale customers data.csv")
x = data["Fresh"]
y = data["Milk"]
data = list(zip(x,y))

# Plot the dataset to identify the clusters
plt.figure()
plt.scatter(x, y, marker='o', facecolors='none', edgecolors='k', s=30)
plt.title('Input data')
plt.xticks(())
plt.yticks(())

# Load the Model
kmeans = KMeans(init='k-means++', n_clusters=3)

# Train the Model
kmeans.fit(data)

# Use the Model to predict values
results = kmeans.predict(data)
print(results)

# Find the centroids
centroids = kmeans.cluster_centers_
print(centroids)

# Plot the clusters along with centroids
plt.figure()
plt.scatter(x, y, marker='o', facecolors='none', edgecolors='k', s=30)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=200, color='r')
plt.title('Input data')
plt.xticks(())
plt.yticks(())
```
