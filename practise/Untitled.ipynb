# 1. Print first 7 records

```python
import pandas as pd

df = pd.read_csv('students.csv')
# 1. Print first 7 records
print("First 7 records:")
df.head(7)
```

```python
#2 Print the Coloumn name
print("\nColumn names:")
print(df.columns.tolist())
```

# 3 Print the statestical information of DataFrame.

```python
print("\nStatistical information:")
df.describe()
```

# 4. Print All Records in Alphabetical Order by Name

```python
print("\nRecords sorted alphabetically by name:")
df.sort_values(by='name')
```

# 5. Print Records Ordered by Highest Marks

```python
# Adding a total_marks column
df['TotalMarks'] = df[['Marks1', 'Marks2', 'Marks3', 'Marks4']].sum(axis=1)

print("\nRecords sorted by highest total marks:")
df.sort_values(by='TotalMarks', ascending=False)
```

# 6. Add Column "Percentage"

```python
# Assuming each exam is out of 100, the percentage is calculated as:
df['Percentage'] = (df[['Marks1', 'Marks2', 'Marks3', 'Marks4']].sum(axis=1) / 4)

print("\nDataFrame with percentage column:")
df
```

# 7. Print Details of Students Living in Rajkot

```python
print("\nStudents living in Rajkot:")
df[df['city'] == 'Rajkot']
```

# 8. Print Students Who Achieve Marks < 35

```python
print("\nStudents who achieve marks < 35:")
# As no records with marks < 35 are in the provided data, this will be empty.
# But here's how you would filter:
df[(df[['Marks1', 'Marks2', 'Marks3', 'Marks4']] < 35).any(axis=1)]
```

# 9. Print Students Whose Name Starts with 'M'

```python
print("\nStudents whose name starts with 'M':")
df[df['name'].str.startswith('M')]
```

# 10. Find Student with Maximum Percentage

```python
max_percentage_student = df[df['Percentage'] == df['Percentage'].max()]
print("\nStudent with maximum percentage:")
max_percentage_student
```

# 11  Find Student with Minimum Percentage

```python
min_percentage_student  = df[df['Percentage'] == df['Percentage'].min()]

print("\nStudent with minimum percentage:")
min_percentage_student 
```
