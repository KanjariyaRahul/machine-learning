```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Plotting all images
plt.figure(figsize=(15, 15))

# 1) Plotting the original image
# Read the original image
img = cv2.imread("3.jpg")
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(1, 3, 1)
plt.imshow(new_img)
plt.title('Original Image')
plt.axis('off')

# 2) Define a kernel for filtering
k1 = np.array([[-1, -1, -1],[-1,  9, -1],[-1, -1, -1]])
# Apply the filter to the original image
filtered_img = cv2.filter2D(img, -1, k1)
# Plotting the filtered image
plt.subplot(1, 3, 2)
plt.imshow(filtered_img)
plt.title('Filtered Image')
plt.axis('off')

#3) blur image 
blurred_img = cv2.blur(new_img, (5,5))
# Plotting the blurred image
plt.subplot(1, 3, 3)
plt.imshow(blurred_img)
plt.title('Blurred Image')
plt.axis('off')

plt.show()
```

# Box, Gaussian, Median Blur


```python
import cv2
import matplotlib.pyplot as plt

plt.figure(figsize=(50, 30))

img= cv2.imread("3.jpg")

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#box blur
blurred_img1 = cv2.blur(img,(8, 8))

#gaussian blur
blurred_img2 = cv2.GaussianBlur(img,(3,3),0)

#median blur
blurred_img3 = cv2.medianBlur(img,5)

#subplot(row, col,sequence)
plt.subplot(1,4,1),plt.imshow(img),plt.title("Original Image")
plt.axis('off')
plt.subplot(1,4,2),plt.imshow(blurred_img1),plt.title("Box Image")
plt.axis('off')
plt.subplot(1,4,3),plt.imshow(blurred_img2),plt.title("Gaussian Image")
plt.axis('off')
plt.subplot(1,4,4),plt.imshow(blurred_img3),plt.title("median Image")
plt.axis('off')
```

# sharpening filters in image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img= cv2.imread("5.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15, 15))

k1 = np.array ([[-1,-1,-1] , [-1, 9, -1] , [-1,-1,-1]])
k2 = np.array ([[1,1,1] , [1, -7, 1] , [1,1,1]])
new_img1 = cv2.filter2D(img , -1 , k1)
new_img2 = cv2.filter2D(img , -1 , k2)
plt.subplot(1,3,1),plt.imshow(img)
plt.axis('off')
plt.subplot(1,3,2),plt.imshow(new_img1)
plt.axis('off')
plt.subplot(1,3,3),plt.imshow(new_img2)
plt.axis('off')
```

# embossing images filter

```python

k1 = np.array ([[0,-1,-1] , [1, 0, -1] , [1,1,0]])
k2 = np.array ([[-1,-1,0] , [-1, 0, 1] , [0,1,1]])

plt.figure(figsize=(15, 15))
new_img1 = cv2.filter2D(img , -1 , k1) +128
new_img2 = cv2.filter2D(img , -1 , k2) + 128

plt.subplot(1,3,1),plt.imshow(img)
plt.axis('off')
plt.subplot(1,3,2),plt.imshow(new_img1)
plt.axis('off')
plt.subplot(1,3,3),plt.imshow(new_img2)
plt.axis('off')
```

# Edge Detection  1) sobel

```python
#edge detection : soble (img,ddepth ,x,y, size)

sobelx= cv2.Sobel(img,cv2.CV_64F , 1,0,5)
sobely= cv2.Sobel(img,cv2.CV_64F , 0,1,5)

laplacian_img = cv2.Laplacian(img, cv2.CV_64F)

canny_img = cv2.Canny(img, 50 , 150 )
plt.figure(figsize=(15, 15))
plt.subplot(1,3,1),plt.imshow(img , cmap="gray")
plt.axis('off')
plt.subplot(1,3,2),plt.imshow(sobelx, cmap="gray")
plt.axis('off')
plt.subplot(1,3,3),plt.imshow(sobely, cmap="gray")
plt.axis('off')

plt.subplot(1,3,4),plt.imshow(laplacian_img, cmap="gray")
plt.axis('off')

plt.subplot(1,3,5),plt.imshow(canny_img, cmap="gray")
plt.axis('off')
```
